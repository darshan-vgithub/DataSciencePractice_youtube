```python
import numpy as np
```

```python
a=np.array([1,2,3,4,5,6],dtype=np.int8) #1D array
print(a)
```

```python
b=np.array([[9.0,8.0,7.0],[6.0,5.0,4.0]]) #2D array
print(b) 
```

```python
# get the dimensions of the arrays
print(a.ndim) #1D array
print(b.ndim) #2D array
```

```python
# get shape of the arrays
print(a.shape) #1D array has shape (6,)
print(b.shape) #2D array has shape (2, 3)
```

```python
# get data type of the arrays
print(a.dtype) #1D array has data type int64
print(b.dtype) #2D array has data type float64
```

```python
# get item size of the array
print(a.itemsize) #1D array has item size 1 byte
print(b.itemsize) #2D array has item size 8 bytes
```

```python
# get total size of the array
print(a.nbytes) #1D array has total size 6
print(b.nbytes) #2D array has total size 48
```
