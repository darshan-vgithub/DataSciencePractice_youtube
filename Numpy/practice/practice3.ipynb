Intializing the diferent types of arrays

```python
import numpy as np
```

```python
a=np.zeros(5)
print(a)
```

```python
b=np.ones((2,3))
print(b)
```

```python
c=np.full((2,3),5)
print(c)
```

```python
d=np.random.rand(2,3)
print(d)
```

```python
e=np.random.randint(1,11,(2,3))
print(e)
```

```python
output=np.ones((5,5))
print(output)

z=np.zeros((3,3))
# print(z)
z[1,1]=9
print(z)

output[1:4,1:4]=z
print(output)
```
